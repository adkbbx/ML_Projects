# Popularity based rs
# books with highest average rating
# minimum 250 users on book rating

```python
Book_ratings = ratings.merge(books,on='ISBN')
```

```python
ratings.merge(books,on='ISBN')
```

```python
Book_ratings
```

```python
num_ratings_df = Book_ratings.groupby("Book-Title").count()["Book-Rating"].reset_index()
num_ratings_df.rename(columns={'Book-Rating':'num-ratings'},inplace=True)
num_ratings_df
```

```python
# Check the data type of 'Book-Rating'
print("Data type of 'Book-Rating':", Book_ratings['Book-Rating'].dtype)

# Check unique values in 'Book-Rating'
unique_ratings = Book_ratings['Book-Rating'].unique()
print("Unique values in 'Book-Rating':", unique_ratings)
```

```python
# Group by "Book-Title" and calculate mean of "Book-Rating"
avg_ratings_df = Book_ratings.groupby("Book-Title").agg({'Book-Rating': 'mean'}).reset_index()
avg_ratings_df.rename(columns={'Book-Rating':'avg-rating'},inplace=True)
avg_ratings_df
```

```python
popularity_df = num_ratings_df.merge(avg_ratings_df,on='Book-Title')
popularity_df
```

```python
popularity_df= popularity_df[popularity_df['num-ratings']>=250].sort_values('avg-rating', ascending=False).head(50)
```
